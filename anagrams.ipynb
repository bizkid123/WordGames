```python
MAXLETTERS = 6
MINLETTERS = 3
def load(filePath):
    lines = []
    with open(filePath) as file:
        while (line := file.readline().strip()):
            lines.append(line)

    lines = [word for word in lines if MAXLETTERS >= len(word) >= MINLETTERS]

    return lines
```

```python
def points(word):
    # return 1
    match len(word):
        case 3:
            return 100
        case 4:
            return 400
        case 5:
            return 1200
        case 6:
            return 2000
        case 7:
            return 3000
def anagramOf(bigWord,smallWord):
    return all(bigWord.count(c) >= smallWord.count(c) for c in smallWord)

def maxScore(s,wordsList):
    # print(len(wordsList))
    return sum(points(word) for word in wordsList if anagramOf(s, word))

print(maxScore("tapers", load("wordsDict.txt")))
```

```python
def hillClimb():
    import random

    alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    words = load("wordsDict.txt")

    bestLetters = "".join(random.sample(alphabet, MAXLETTERS))

    bestScore = maxScore(bestLetters, words)

    for _ in range(1000):
        newChar = random.choice(alphabet)
        charPos = random.choice([i for i in range(MAXLETTERS)])

        newLetters = list(bestLetters)
        newLetters[charPos] = newChar
        newLetters = ''.join(newLetters)

        newScore = maxScore(newLetters, words)

        if newScore > bestScore:
            bestScore = newScore
            bestLetters = newLetters
            print(bestLetters, bestScore)
    return bestLetters, bestScore
hillClimb()
```

```python
def bruteForce():
    words = load("wordsDict.txt")
    uniqueWords = set([''.join(sorted(word)) for word in words if len(word) == MAXLETTERS])
    print(len(uniqueWords))
    top = 10
    best = [('',0)]*top
    for i,word in enumerate(uniqueWords):
        if i%500 == 0:
            print(best, i)
        score = maxScore(word, words)
        if score > best[-1][1]:
            best.pop()
            best.append((word, score))
            best.sort(key = lambda x: x[1], reverse = True)
    return (best)
# bruteForce()
```
